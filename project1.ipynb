Apply method 

```python
def grade(x):
    if x == 'A':
        return "Superb"
    elif x == 'B':
        return "good"
    elif x == 'C':
        return "not bad"
    else:
        return "better luck next time"
        
```

```python
data['review'] = data["Grade"].apply(grade)
print(data['review'])
```

encoding using map function



```python
data['male_female'] = data['Gender'].map({"male":1,"female":0})
data
```

```python
data.drop(['male_female','review'],axis= 1 ,inplace= True)
```
